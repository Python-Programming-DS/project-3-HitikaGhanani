# **Part 1: Data Cleaning**

Reading and Examining Dataset


```python
#importing necessary libraries
import pandas as pd
import numpy as np
```

```python
#Loading data
df= pd.read_csv("/content/new_train_EGN5442.csv")
print(df)
```

```python
df.info()
```

```python
df.describe()
```

Handling columns which are not numeric: x2, x3, x14, x25, x26, x29, x30, x31


```python
#Cleaning dataset
print(df[["x2", "x3", "x14", "x25", "x26", "x29", "x30","x31"]])

# x2 and x29 look like they represent some id or something else which cannot be decoded. So we will drop these columns
df= df.drop(columns=['x2','x29'])

# column x30 has NAN values and does not provide any information so we will drop it
df= df.drop(columns=['x30'])

```

```python
# handling x3, x25 and x26
print(df['x3'].unique())
print(df['x25'].unique())
print(df['x26'].unique())

# These three columns can be coded into binary
df['x3'] = df['x3'].map({'Female': 0, 'Male': 1})
df['x25'] = df['x25'].map({'D_C': 0, 'L_C': 1})
df['x26'] = df['x26'].map({'Bt': 0, 'PT': 1})

# Handling inf values in column x31
df['x31'] = df['x31'].replace([np.inf, -np.inf], np.nan)

print(df)
```

```python
# Handling x14 column
df['x14'].unique()

#It seems numeric data but just showing dtype as object
df['x14'] = df['x14'].replace(' ', np.nan)
df['x14'] = df['x14'].astype('Int64')

df.info()
```

```python
#Handling missing values
df.isnull().sum()
```

```python
# Filling missing values with median
df['x14'] = df['x14'].fillna(df['x14'].median())
df['x24'] = df['x24'].fillna(df['x24'].median())
df['x31'] = df['x31'].fillna(df['x31'].median())

print(df)

```

```python
# Checking for duplicated rows
df[df.duplicated()]
df = df.drop_duplicates(keep='first')

```

```python
# Checking for highly correlated columns/features
corr = df.corr(numeric_only=True)
threshold = 0.9
high_corr = np.where(np.abs(corr) > threshold)
high_corr_pairs = [(corr.index[i], corr.columns[j], corr.values[i, j])
    for i, j in zip(*high_corr)
    if i < j]

high_corr_pairs
```

```python
df[['x5', 'x6', 'x23', 'x24']]
```

```python
# As column x5 and x6 are identical and
# column x23 and x24 are highly correlated we may drop each one of them

df = df.drop(['x5', 'x24'], axis=1)

df.info()

#As we can see there is ni categorical column left because it was handled earlier only by binary encoding
```

```python
# Sorting into bins
df.nunique().sort_values()
```

```python
# We can see only x6, x23 and x31 have more number of unique values so we will only sort them into bins

cols_bin = df[['x6', 'x23', 'x31']]
df_binned = df.copy()
for col in cols_bin:
    df_binned[col + '_bin'] = pd.qcut(df[col], q=10, labels=False, duplicates='drop')
```

```python
df_binned[['x6', 'x6_bin', 'x23', 'x23_bin', 'x31', 'x31_bin']].describe()
```

```python
df_binned = df_binned.drop(columns=cols_bin)
df_binned.info()
```

```python
# Visualisations
import matplotlib.pyplot as plt
import seaborn as sns

# Heat Map
plt.figure(figsize=(14, 12))
sns.heatmap(df_binned.corr(numeric_only=True), cmap='coolwarm', center=0)
plt.title("Correlation Heatmap of Features")
plt.show()

```

```python
# To check if the classes are imbalanced
sns.countplot(x=df_binned['y'])
plt.title("Distribution of Target Variable y")
plt.show()
```

```python
# Distribution plots
ordinal_cols = [col for col in df_binned.columns if df_binned[col].nunique() <= 7 and col != 'y']

plt.figure(figsize=(16, 14))
for i, col in enumerate(ordinal_cols[:12]):
    plt.subplot(4, 3, i+1)
    sns.countplot(x=df[col])
    plt.title(f'Distribution of {col}')
plt.tight_layout()
plt.show()
```
